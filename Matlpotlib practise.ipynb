```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)   #use plt.show() to print the graph
plt.xlabel('Xlabel')
plt.ylabel('ylabel')
plt.title('title')
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

Object Oriented method

```python
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('t')
```

```python
fig=plt.figure()
a1=fig.add_axes([0.1,0.1,0.8,0.8])
a1.set_title('Big plot')
a2=fig.add_axes([0.2,0.4,0.3,0.4])
a2.set_title('Small plot')
a1.plot(x,y,'b*-')
a2.plot(y,x,'r.-')
```

```python
f,ax = plt.subplots(nrows=1,ncols=3)
```

```python
i=0
for a in ax:
    a.set_title('title'+str(i))
    i=int(i)
    i=i+1
    a.plot(x,y)
f
```

```python
ax[1].set_xlabel('x1')
ax[1].set_ylabel('y1')
ax[1].plot(y,x)

plt.tight_layout()
f
```

```python
f=plt.figure()
a=f.add_axes([0,0,1,1])
a.plot(x,y, label='x square')
a.plot(x,x**3, label='x cube')
a.legend()
```
